```python
a = tf.Variable(np.random.uniform(0, 1, (2, 3)))
b = preprocess_input(a*255.)
with tf.GradientTape() as tape:
    tape.watch(a)
    b = a*10.
    print(b)
    c = b 
    print(c)
gradients = tape.gradient(c, a)
print(gradients)

print(tf.clip_by_value(gradients, 0., 1.))
```
